```python
np.random.seed(42)
num_rows = 1000
products = ['Розы', 'Ромашки', 'Гипсофилы', 'Хризантемы', 'Тюльпаны', 'Орхидеи', 'Лилии', 'Гортензии', 'Плюшевый мишка', 'Конфеты', 'Торт', 'Шары', 'Плюшевый заяц', 'Пирожные', 'Конфетти']
categories = ['Цветы', 'Цветы','Цветы','Цветы','Цветы','Цветы','Цветы','Цветы', 'Подарки', 'Сладости', 'Сладости', 'Подарки', 'Подарки','Сладости','Подарки']
price_ranges = ['250', '150', '100', '500', '150', '700', '300', '500', '2000', '300', '1000', '100', '3000', '150', '100']
product_category = dict(zip(products, categories))
product_price = dict(zip(products, price_ranges))
data = {
    'order_id': range(1000, 1000 + num_rows),
    'products': np.random.choice(products, num_rows),
    'quantity': np.random.randint(1, 100, num_rows),
    'order_date': pd.date_range(start='2025-01-01', end='2025-12-31', periods=num_rows)
}
data['category'] = [product_category[product] for product in data['products']]
data['price'] = [float(product_price[product]) for product in data['products']]

df2 = pd.DataFrame(data)
```

```python

np.random.seed(42)
n = 300

df2 = pd.DataFrame({
    'age': np.random.randint(18, 70, n),
    'income': np.round(np.random.uniform(20000, 120000, n), 2),
    'education': np.random.choice([
        'Неполное среднее', 'Среднее', 'Среднее специальное',
        'Бакалавр', 'Магистр', 'Кандидат наук'
    ], n),
    'city': np.random.choice(['Москва', 'Санкт-Петербург', 'Новосибирск', 'Красноярск'], n)
})

print("Исходные данные (первые 3 строки):")
print(df2.head(3))
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

le_edu = LabelEncoder()
df['education_encoded'] = le_edu.fit_transform(df['education'])

mapping = pd.DataFrame({
    'education': le_edu.classes_,
    'encoded': le_edu.transform(le_edu.classes_)
})
print('образование через LabelEncoder:')
print(mapping)
print()

ohe = OneHotEncoder(sparse_output=False, drop='first')
nominal_encoded = ohe.fit_transform(df[['city']])
nominal_cols = ohe.get_feature_names_out(['city'])
df_nominal = pd.DataFrame(nominal_encoded, columns=nominal_cols, index=df.index)

scaler = StandardScaler()
df[['age', 'income']] = scaler.fit_transform(df[['age', 'income']])

df_final = pd.concat([
    df[['age', 'income', 'education_encoded']],df_nominal], axis=1)

print('получился вот такой датасет:')
print(df_final.head(3))
```
