```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/fitness_data.csv')
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

Датасет содержит подробную информацию о личных качествах, питании и физической форме 20000 человек. Он предназначен для анализа того, как рацион питания, физические нагрузки и состав тела влияют на расход калорий, уровень физической подготовки и общее состояние здоровья.

```python
sns.set_theme(style="whitegrid", font_scale=1)
plt.figure(figsize=(10,5))

sns.lineplot(data=df, x="Session_Duration (hours)", y="Avg_BPM", linewidth=3)

plt.title("Средний пульс в зависимости от длительности тренировки")
plt.xlabel("Длительность тренировки (часы)")
plt.ylabel("Средний пульс (уд/мин)")
plt.tight_layout()
plt.show()
```

```python
ax1=sns.lineplot(data=df, x="Session_Duration (hours)", y="Calories_Burned", hue="Workout_Type", linewidth=2.5, palette="flare")
plt.title("Калории в зависимости от длительности тренировки с учётом типа тренировки")
plt.xlabel("Длительность тренировки (часы)")
plt.ylabel("Калории")
plt.tight_layout()
ax1.legend_.set_title("Тип тренировки")
plt.show()
```

```python
plt.figure(figsize=(15,7))
ax2 = sns.histplot(data=df, x='BMI', hue='Gender', kde=True, palette={"Male": "#00b3ff", "Female": "#ff00d5"}, alpha=0.6)
ax2.legend(labels=['Женщины', 'Мужчины'])
plt.title("Распределение ИМТ по полу")
plt.xlabel("Индекс массы тела")
plt.ylabel("Количество")
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax3= sns.histplot(data=df, x='cal_balance', hue='Burns_Calories_Bin', multiple='stack', palette='viridis')
plt.title("Распределение калорийного баланса по категориям интенсивности тренировки")
plt.xlabel("Баланс калорий")
plt.ylabel("Количество")
legend = ax3.legend(labels=['Средняя', 'Высокая', 'Низкая', 'Очень высокая'])
legend.set_title("Интенсивность")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
ax4 = sns.scatterplot(data=df, x='protein_per_kg', y='Fat_Percentage', hue='Experience_Level', size='Experience_Level', palette='coolwarm')
legend = ax4.legend()
plt.title('Связь между потреблением белка и процентом жира в организме по уровню фитнес-опыта')
plt.xlabel("Белок на кг массы тела (г/кг)")
plt.ylabel("Процент жира в организме")
legend.set_title("Уровень опыта занятий фитнесом")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
ax5 = sns.scatterplot(data=df,x='Workout_Frequency (days/week)', y='cal_balance', hue='Burns_Calories_Bin', s=20)
legend = ax5.legend(labels=['Средний', 'Высокий', 'Низкий', 'Очень высокий'])
plt.title('Связь между частотой тренировок и калорийным балансом')
plt.xlabel("Частота тренировок (дней в неделю)")
plt.ylabel("Баланс калорий")
legend.set_title("Уровень интенсивности сжигания\nкалорий на тренировке")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
ax6 = sns.scatterplot(data=df, x="Calories", y="Fat_Percentage", hue="Age", s=50, alpha=0.6)
plt.title("Связь между суточным потреблением калорий и процентом жира в организме")
plt.xlabel("Суточное потребление калорий (ккал)")
plt.ylabel("Процент жира в организме")
ax6.legend_.set_title("Возраст")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.boxenplot(data=df, x="Workout_Type", y="Fat_Percentage", palette="flare", hue='Workout_Type', legend=False)
plt.title("Распределение процента жира в организме в зависимости от типа тренировки", fontsize=15)
plt.xlabel("Тип тренировки")
plt.ylabel("Процент жира в организме")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
df['Exp'] = pd.cut(df['Experience_Level'], bins=3, labels=['Новичок', 'Средний уровень', 'Опытный'])
ax = sns.boxenplot(data=df, x="Exp", y="cal_balance", palette="muted", hue = 'Exp', legend=False)
plt.title("Калорийный баланс в зависимости от уровня фитнес-опыта")
plt.xlabel("Уровень опыта")
plt.ylabel("Калорийный баланс (ккал)")
plt.axhline(0, color='gray', linestyle='--', linewidth=1, label='Нулевой баланс')
plt.legend()
plt.tight_layout()
plt.show()
```
